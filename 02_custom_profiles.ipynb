## Working with custom CIM profiles

In addition to the pre-built profiles shown in Example 1, CIMantic Graphs supports **custom profiles** authored in tools like CIMTool or SchemaComposer. A custom profile is a trimmed-down subset of the CIM schema containing only the classes and attributes your application needs.

This example uses `CoreEquipment`, a custom profile bundled with the demo. It is imported exactly like a built-in profile.

```python
import os
import cimgraph_demo.custom_profiles.CoreEquipment as cim
os.environ['CIMG_CIM_PROFILE'] = 'cimgraph_demo.custom_profiles.CoreEquipment'
```

Compared with the 1,174 classes in the full CIM 17 profile, this custom profile is far smaller — containing only the classes selected when it was authored:

```python
print(f'This profile contains {len(cim.__all__)} classes')
```

### Declared vs. inherited attributes

Each profile class is a Python dataclass. The `__annotations__` of a class list only the attributes **declared directly on that class**, while `__dataclass_fields__` includes every field, including those **inherited** from parent classes.

For `ACLineSegment`, this distinction shows how many attributes are specific to the line segment itself versus inherited from its CIM ancestors (e.g. `Conductor`, `ConductingEquipment`, `Equipment`).

```python
print(f'The Class ACLineSegment has {len(cim.ACLineSegment.__annotations__)} attributes')
```

```python
print(f'The Class ACLineSegment has {len(cim.ACLineSegment.__dataclass_fields__)} total attributes')
```

## Summary

Custom profiles behave just like the built-in ones: you import the module, point `CIMG_CIM_PROFILE` at it, and work with the generated dataclasses. The only difference is scope — a custom profile exposes just the classes and attributes you chose to include.

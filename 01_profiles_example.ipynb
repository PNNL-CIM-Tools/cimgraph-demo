## Working with pre-built CIM-Graph Profiles

This example shows how to load a pre-built CIM Profile and work with the dataclasses within that profile

```python
import os
import cimgraph.data_profile.cim17v40 as cim
os.environ['CIMG_CIM_PROFILE'] = 'cim17v40'
```

```python
print(f'This profile contains {len(cim.__all__)} classes')
```

```python
import cimgraph.data_profile.cim18gmdm.connectivity as cim
os.environ['CIMG_CIM_PROFILE'] = 'cimgraph.data_profile.cim18gmdm.connectivity'
print(f'This profile contains {len(cim.__all__)} classes')
```

```python
import cimgraph.data_profile.cimhub_2023 as cim
print(f'This profile contains {len(cim.__all__)} classes')
```

```python
from mermaid import Mermaid
from cimgraph.utils import get_mermaid
Mermaid(get_mermaid(cim.EnergyConsumer))
```

```python
print(cim.EnergyConsumer.__doc__)
```

```python
diagram = get_mermaid([cim.EnergyConsumer, cim.EnergyConsumerPhase])
Mermaid(diagram)
```

```python
from cimgraph.utils import add_mermaid_path
diagram = get_mermaid(cim.ACLineSegment)
diagram = add_mermaid_path(cim.ACLineSegmentPhase, 'WireInfo', diagram)
Mermaid(diagram)
```
